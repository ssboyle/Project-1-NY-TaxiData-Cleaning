```python
import pandas as pd
import numpy as np
import csv
```

```python
url = "../../Project_TaxiData/feb2015_df2.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# splitting the pickup_datetime column into date and time
datetime = df["tpep_pickup_datetime"].str.split(" ", n = 1, expand = True)
```

```python
datetime.head()
```

```python
# creating lists for the split data
time = list(datetime[1])
date = list(datetime[0])
vendor = list(df["VendorID"])
```

```python
# creating clean new dataframe
listdata = {
    "Date": date,
    "Time": time,
    "Taxi_Vendor": vendor
    
}

clean_df = pd.DataFrame(listdata)

clean_df.head()
```

```python
group = clean_df.groupby(["Date"])
group.head()
```

```python
count_df = group.count()
count_df.head()
```

```python
count_df.to_csv("feb2015summary.csv")
```
